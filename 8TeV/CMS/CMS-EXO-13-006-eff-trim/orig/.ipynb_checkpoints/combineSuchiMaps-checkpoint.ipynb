```python
import numpy as np
import glob,os
```

```python
massDims = {'THSCPM1b' : 1, 'THSCPM2' : 1, 'THSCPM3' : 2, 'THSCPM4' : 2, 'THSCPM5' : 3, 'THSCPM6' : 3, 
            'THSCPM7' : 3, 'THSCPM8' : 2, 'THSCPM9' : 2}

def mergeFolder(folder):
    mapData = None
    topo = folder.split('efficiencies_')[1]
    for f in glob.glob(os.path.join(folder,'*.txt')):        
        mDim = massDims[topo]
        if mDim == 1:
            massLabels = ['mhscp']
        elif mDim == 2:
            massLabels = ['mprod','mhscp']
        else:
            massLabels = ['mprod','mint','mhscp']
        columnLabels = massLabels + ['width','c000','c100','c200','c300']
#         print(f)
        data = np.genfromtxt(f)
        if len(data) == 0:
            continue
        data = data[:,-len(columnLabels):]
        if mapData is None:
            mapData = data
        else:
            mapData = np.vstack((mapData,data))
            
    mapDataF = np.array([tuple(pt) for pt in mapData],dtype=[(c,float) for c in columnLabels])
    mapDataF = np.sort(mapDataF,order=massLabels+['width'])
    #Save results to file:
    outputFile = '%s_efficiencyMaps.txt' %topo
    header = '%19s'*len(mapDataF.dtype.names) %mapDataF.dtype.names
    header = header[3:]
    np.savetxt(outputFile,mapDataF,header=header,fmt = ['     %1.7e']*len(mapDataF.dtype.names))
    return outputFile
```

```python
folders = ['./efficiencies_THSCPM1b','./efficiencies_THSCPM2','./efficiencies_THSCPM3',
           './efficiencies_THSCPM4','./efficiencies_THSCPM6','./efficiencies_THSCPM8',
          './efficiencies_THSCPM9']
```

```python
for f in folders:
    outF = mergeFolder(f)
    print('Folder %s merged to %s' %(f,outF))
```

```python
mergeFolder('./efficiencies_THSCPM5')
```
