```python
import matplotlib.pyplot as plt
import os,glob
import numpy as np
import seaborn as sns
from scipy.special import kn,zetac
import copy
import mypyslha as pyslha
from checkMCstats import checkStats
sns.set() #Set style
sns.set_style('ticks',{'font.family':'serif', 'font.serif':'Times New Roman'})
sns.set_context('paper', font_scale=1.8)
sns.set_palette(sns.color_palette("Paired"))
```

```python
MCstats = []
for f in glob.glob('../data/checkmate_BM2/*/evaluation/total_results.txt'):
    massTag = f.replace('../data/checkmate_BM2/','').split('/')[0].split('_')
    mchi1 = eval(massTag[1])
    mchi0 = eval(massTag[3])
    r, nevts = checkStats(f)
    MCstats.append([mchi1,mchi0,r,nevts])
MCstats = np.array(MCstats)
```

```python
cp = plt.scatter(MCstats[:,0],MCstats[:,0]-MCstats[:,1],c=MCstats[:,2],s=35,
                 cmap=plt.cm.get_cmap('RdYlBu'), vmin=0., vmax=20.)
cb = plt.colorbar(cp)
cb.set_label(r'$N_{MC}/N_{signal}$')
plt.xlabel(r'$m_{\chi_1}$ (GeV)')    
plt.ylabel(r'$m_{\chi_1}-m_{\chi_0}$ (GeV)')
plt.show()
```

```python
cp = plt.scatter(MCstats[:,0],MCstats[:,0]-MCstats[:,1],c=MCstats[:,3],s=35,
                 cmap=plt.cm.get_cmap('RdYlBu'), vmin=10., vmax=100.)
cb = plt.colorbar(cp)
cb.set_label(r'$N_{MC}$')
plt.xlabel(r'$m_{\chi_1}$ (GeV)')    
plt.ylabel(r'$m_{\chi_1}-m_{\chi_0}$ (GeV)')
plt.show()
```
